```python
# PATH FOR ALL STATES IN AGGREGATED TRANSACTIONS
path = r"C:\Users\HP\Desktop\phonepe\pulse\data\aggregated\transaction\country\india\state/"
st_list = os.listdir(path)
st_list
```

```python
# INITIALIZATION
aff_df = {'State':[],'Year' : [],'Quarter': [], 'Transaction_type':[], 'Total_transaction':[],'Transaction_amount':[]}

for i in st_list:
  p1 = path + i +'/'
  agg_t_year = os.listdir(p1)

  for j in agg_t_year:
    p2 =  p1 +j + '/'
    agg_t_quarter = os.listdir(p2)

    for k in agg_t_quarter:
      p3 = p2 + k
      data = open(p3,'r')
      d = json.load(data)
      for z in d['data']['transactionData']:
        Name = z['name']
        Count = z['paymentInstruments'][0]['count']
        Amount = z['paymentInstruments'][0]['amount']
        aff_df['State'].append(i)
        aff_df['Year'].append(j)
        aff_df['Quarter'].append(int(k.strip('.json')))
        aff_df['Transaction_type'].append(Name)
        aff_df['Total_transaction'].append(Count)
        aff_df['Transaction_amount'].append(Amount)

aff_trans = pd.DataFrame(aff_df)
aff_trans
```
